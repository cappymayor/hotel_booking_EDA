```python
import pandas as pd
```

```python
df = pd.read_csv("hotel_bookings.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#filling all missing values with 0, dropping all will shrink our dataset, so we assign 0 since we will compute for number of children later.
df.fillna(0, inplace = True)
```

```python
#confirming if all missing values are replced, now we have no missing values.
df.isnull().sum()
```

```python
#checking for the hotel with the highest patronage in the Dataset.
df.groupby("hotel", as_index = False)["customer_type"].count().rename(columns = {"customer_type":"Total"})
```

```python
#checking the hotel with the highest number of children
df.groupby("hotel", as_index = False)["children"].sum().rename(columns = {"children":"Total_Children"})
```

```python
#checking the hotel with the highest number of adults
df.groupby("hotel", as_index = False)["adults"].sum().rename(columns = {"adults":"Total_Adult"})
```

```python
df.groupby("hotel", as_index = False)["country"].count().rename(columns = {"country":"Country_Total"})
```

```python
df.groupby(["hotel","arrival_date_year"], as_index = False)["customer_type"].count().rename(columns = {"customer_type":"Total"})
```

```python
#figuring the peak month for bookings among both hotels.
df.groupby(["hotel","arrival_date_month"],as_index = False)["customer_type"].count().sort_values(by = "customer_type", ascending = False).rename(columns = {"customer_type":"Total"})
```

```python
#looking at the top 5 week of the year with the highest booking 
peak_weak = df.groupby("arrival_date_week_number", as_index = False)["customer_type"].count()
peak_weak.sort_values(by = "customer_type", ascending=False).head().reset_index(drop = True).rename(columns = {"customer_type":"Total"})
```

```python
df.groupby(["is_repeated_guest", "hotel"], as_index = False)["customer_type"].count().rename(columns = {"customer_type":"Total"})
```

```python
df.groupby(["hotel", "customer_type"], as_index = False)["company"].count().rename(columns = {"company":"Total"})
```

# SPLITTING OUR DATA INTO TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size = .2, random_state = 10)
```

```python
train_X = train[categorical_columns + numerical_columns].reset_index(drop=True)
train_y = train[target_column].reset_index(drop=True)

test_X = test[categorical_columns + numerical_columns].reset_index(drop=True)
test_y = test[target_column].reset_index(drop=True)
```
